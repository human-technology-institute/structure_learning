```python
import numpy as np
from mcmc.tests.test_simulation_structure_mcmc_synthetic import simulate

N_RESTARTS = 100
```

```python
# small graphs (n=5, d=2)
n = 5
results = [r for r in simulate(n_restarts=N_RESTARTS, n_nodes=n, node_degree=2, n_iterations=30000, save_results=True, results_dir=f'nodes_{n}', seed=46)]

# percentage of simulations in which the true DAG was found
is_true_dag_found = [r[3] for r in results]
print(np.count_nonzero(is_true_dag_found)/len(is_true_dag_found)*100)
```
